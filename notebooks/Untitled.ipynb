```python
numeric_df = df3.select_dtypes(include='number')
negative_columns = numeric_df.columns[(numeric_df < 0).any()]
print("Columns with negative values Are:", negative_columns)

negative_counts = (numeric_df < 0).sum()
print("Sum of negative values in each numeric column:")
print(negative_counts)
```

```python
dataframes = [df1, df2, df3]

df_names = ['Benin_Data', 'sieralion_data', 'togo_data']

# Loop through each dataframe
for idx, df in enumerate(dataframes):
    print(f"\n{df_names[idx]}: Negative Values Description")
    
    for column in ['GHI', 'DNI', 'DHI']:
        print(f"\nColumn: {column}")
        if column in df.columns:
            print(df[df[column] < 0][column].describe())
        else:
            print(f"{column} does not exist in {df_names[idx]}")
```

```python
# Replace negative values with NaN (for later imputation or removal)
df3['GHI'] = df3['GHI'].apply(lambda x: x if x >= 0 else None)

# Example of replacing negative values with the mean of positive values
mean_value = df3[df3['GHI'] > 0]['GHI'].mean()
df3['GHI'].fillna(mean_value, inplace=True)
```

```python

# Set 'Date' as the index
df1.set_index('Timestamp', inplace=True)

# Plot an area plot
df1.plot(kind='area', stacked=True, alpha=0.5)

# Customize the plot
plt.title('Area Plot Over Time')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend(title='Categories')
plt.show()
```
